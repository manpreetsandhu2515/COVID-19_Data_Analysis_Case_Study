### Data cleaning and handling missing values 

```python
deaths_long['Province/State'] = deaths_long['Province/State'].fillna('All Provinces')
deaths_long['Lat'] = deaths_long['Lat'].fillna(0)
deaths_long['Long'] = deaths_long['Long'].fillna(0)
deaths_long['Date'] = pd.to_datetime(deaths_long['Date'], format = "%m/%d/%y", errors = 'coerce')
deaths_long.info()
```

```python
# geo_data = pd.read_csv('https://developers.google.com/public-data/docs/canonical/countries_csv')
```

```python
confirmed_long['Province/State'] = confirmed_long['Province/State'].fillna('All Provinces')
confirmed_long['Lat'] = confirmed_long['Lat'].fillna(0)
confirmed_long['Long'] = confirmed_long['Long'].fillna(0)
confirmed_long['Date'] = pd.to_datetime(confirmed_long['Date'], format = "%m/%d/%y", errors = 'coerce')
confirmed_long.info()
```

```python
recovered_long['Province/State'] = recovered_long['Province/State'].fillna('All Provinces')
recovered_long['Lat'] = recovered_long['Lat'].fillna(0)
recovered_long['Long'] = recovered_long['Long'].fillna(0)
recovered_long['Date'] = pd.to_datetime(recovered_long['Date'], format = "%m/%d/%y", errors = 'coerce')
recovered_long.info()
```

```python
recovered_long.head()
```

```python
recovered_long.sort_values(by = ['Country/Region', 'Date'])
```

```python
deaths_agg = deaths_long.groupby(['Country/Region', 'Date'])['Deaths'].sum().reset_index()
deaths_agg
```

```python
confirmed_agg = confirmed_long.groupby(['Country/Region', 'Date'])['Confirmed'].sum().reset_index()
confirmed_agg
```

```python
recovered_agg = recovered_long.groupby(['Country/Region', 'Date'])['Recovered'].sum().reset_index()
recovered_agg
```

```python
combined_df = (deaths_agg
               .merge(confirmed_agg, on = ['Country/Region', 'Date'], how = 'inner')
               .merge(recovered_agg, on = ['Country/Region', 'Date'], how = 'inner')
               .sort_values(['Country/Region', 'Date'])
              )
combined_df
```

```python
# Ideal approach even if the maximum date was different for countries
# this will give us the index at a country level where the date is maximum 
top_5_countries = confirmed_agg.loc[confirmed_agg.groupby('Country/Region')['Date'].idxmax()].nlargest(5, 'Confirmed')
top_5_countries
```

```python
country_list = top_5_countries['Country/Region'].tolist()
country_list
```

```python
import matplotlib.pyplot as plt
```

```python
#Visualization over time for top countries for confirmed cases
plt.figure(figsize = (14,5))
for country in country_list: 
    top_countries = combined_df[combined_df['Country/Region'] == country].sort_values('Date')
    plt.plot(top_countries['Date'], top_countries['Confirmed'], label = country)
plt.title('Total Confirmed Cases over time for top 5 countries', color = 'red')
plt.xlabel('Dates')
plt.ylabel('Cumulative Confirmed Cases') 
plt.legend() 
plt.show()
```

```python
china_df = confirmed_long[confirmed_long['Country/Region'] == 'China'].sort_values('Date').copy()
china_df
```

```python
china_df = china_df.groupby(['Province/State', 'Date'])['Confirmed'].sum().reset_index() 
china_df
```

```python
china_wide = china_df.pivot(index = 'Date', columns = 'Province/State', values = 'Confirmed').sort_index()
china_wide
```

```python
top_prov = china_wide.iloc[-1].nlargest(5)
top_prov
```

```python
china_wide = china_wide[top_prov.index]
china_wide
```

```python
#Visualization for top 5 provinces of China 
ax = china_wide.plot(figsize = (14,5))
ax.set_title('Confirmed cases over time for top 5 China Provinces')
ax.set_xlabel('Dates') 
ax.set_ylabel('Cumulative Confirmed Cases')
ax.legend(top_prov.index)
plt.show()
```

```python
#5.1  Analyze the peak number of new cases in Germany, France and Italy. 
# Which country experienced the highest single day surge, and when did it occur?
```

```python
target_countries = ['France', 'Germany', 'Italy']
```

```python
confirmed_cases = confirmed_agg[confirmed_agg['Country/Region'].isin(target_countries)]
confirmed_cases
```

```python
confirmed_cases = confirmed_cases.groupby(['Country/Region', 'Date'])['Confirmed'].sum().reset_index() 
confirmed_cases
```

```python
# confirmed_cases.to_excel('confimred_cases.xlsx')
```

```python
confirmed_cases['Lag']  = confirmed_cases.groupby('Country/Region')['Confirmed'].shift(1)
confirmed_cases
```

```python
confirmed_cases['Lag'] = confirmed_cases['Lag'].fillna(0)
confirmed_cases
```

```python
confirmed_cases['Daily New'] = confirmed_cases['Confirmed'] -  confirmed_cases['Lag'] 
confirmed_cases
```
